```python
# Como ler um arquivo do Minio
def list_jsons_from_minio(bucket_name, prefix):
    try:
        # Lista os objetos que começam com o prefixo
        objects = minio_client.list_objects(bucket_name, prefix=prefix, recursive=True)
        
        # Filtra e lê os arquivos que são JSON
        json_files = []
        for obj in objects:
            if obj.object_name.endswith(".json"):  # Filtra apenas arquivos .json
                response = minio_client.get_object(bucket_name, obj.object_name)
                data = json.loads(response.read().decode('utf-8'))
                json_files.append(data)
                response.close()
                response.release_conn()
        return json_files
    except S3Error as e:
        print(f"Erro ao acessar o Minio: {e}")
        return []
```
